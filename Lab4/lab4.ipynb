Метою цiєї лабораторної роботи є отримання навичок роботи з морфологi- чними операцiями над бiнарними зображеннями.

```python
import numpy as np
from PIL import Image, ImageShow
import os
import datetime as dtime
```

```python
def eq(ar1, ar2):
    a1, a2, a3 = ar1
    b1, b2, b3 = ar2
    
    lim1 = (a1 and b1 or not b1)
    lim2 = (a2 and b2 or not b2)
    lim3 = (a3 and b3 or not b3)
    
    if not lim1 or not lim2 or not lim3:
        return False
    return True
```

```python
def h_eq(ar1, ar2):
    a1, a2, a3 = ar1
    b1, b2, b3 = ar2
    
    lim1 = (a1 and b1 or not b1)
    lim2 = (a2 and b2 or not b2)
    lim3 = (a3 and b3 or not b3)
    
    if lim1 or lim2 or lim3:
        return True
    return False
```

```python
def fit(ar3x3):
    a = np.equal(1, fl)
    b = np.equal(0, ar3x3)
    result = []
    
    if not np.any(a[:, 0]):
        pass
    else:
        result.append(eq(b[:, 0], a[:, 0]))
    
    if not np.any(a[:, 1]):
        pass
    else:
        result.append(eq(b[:, 1], a[:, 1]))
        
    if not np.any(a[:, 2]):
        pass
    else:
        result.append(eq(b[:, 2], a[:, 2]))
    
    if np.all(result):
        return True
    return False
```

```python
def hit(ar3x3):
    a = np.equal(1, fl)
    b = np.equal(0, ar3x3)
    result = []
    
    if not np.any(a[:, 0]):
        pass
    else:
        result.append(h_eq(b[:, 0], a[:, 0]))
    
    if not np.any(a[:, 1]):
        pass
    else:
        result.append(h_eq(b[:, 1], a[:, 1]))
        
    if not np.any(a[:, 2]):
        pass
    else:
        result.append(h_eq(b[:, 2], a[:, 2]))
    
    if np.all(result):
        return True
    return False
```

Написати програмну реалiзацiю наступних морфологiчних операцiй: 

1. Erosion (ерозiя).

2. Dilation (нарощування).

3. Opening (розмикання).

4. Closing (замикання).

5. Границi.


```python
def do_morph(photo, fl, fl_type, wild_point_loc, not_rgb=False):
    ar_to_return = photo.copy()
    try:
        height, width, channels = photo.shape
    except ValueError:
        height, width = photo.shape
        not_rgb = True
        channels = 1
    # Ерозія
    if fl_type == 'fit':  
        for channel in range(channels):
            for row in range(1, height - 1):
                for col in range(1, width - 1):
                    if not_rgb:
                        ar_to_return[row, col] = 0 if fit(photo[row-1:row+2, col-1:col+2]) else 255
                    else:
                        ar_to_return[row, col, channel] = 0 if fit(photo[row-1:row+2, col-1:col+2, channel]) else 255
    # Нарощування
    elif fl_type == 'hit':  
        for channel in range(channels):
            for row in range(1, height - 1):
                for col in range(1, width - 1):
                    if not_rgb:
                        ar_to_return[row, col] = 0 if hit(photo[row-1:row+2, col-1:col+2]) else 255
                    else:
                        ar_to_return[row, col, channel] = 0 if hit(photo[row-1:row+2, col-1:col+2, channel]) else 255
    # Відкриття = Ерозія + Нарощування
    elif fl_type == 'open':  
        ar_to_return = do_morph(photo, fl, 'fit', wild_point_loc, not_rgb)
        ar_to_return = do_morph(ar_to_return, fl, 'hit', wild_point_loc, not_rgb)
    # Закриття = Нарощування + Ерозія
    elif fl_type == 'close':  
        ar_to_return = do_morph(photo, fl, 'hit', wild_point_loc, not_rgb)
        ar_to_return = do_morph(ar_to_return, fl, 'fit', wild_point_loc, not_rgb)
    # Границі = Нарощування - Ерозія
    elif fl_type == 'boundary':  
        dilated = do_morph(photo, fl, 'hit', wild_point_loc, not_rgb)
        eroded = do_morph(photo, fl, 'fit', wild_point_loc, not_rgb)
        ar_to_return = dilated - eroded

    return ar_to_return
```

```python
photo_or = Image.open("/Users/yurab/Desktop/IgromanYTS-basics-of-computer-vision/Lab4/images.jpeg")
photo = np.array(photo_or)
print("Photo shape:", photo.shape)
print("Photo unique values:", np.unique(photo))

fl = np.array([
    [0, 1, 0],
    [0, 1, 0],
    [0, 1, 0],
])
```

```python
wild_point_loc = (0, 1)
fl_types = ['fit', 'hit', 'open', 'close', 'boundary']

os.makedirs('Results', exist_ok=True)

for fl_type in fl_types:
    print(f"Processing {fl_type}...")
    result = do_morph(photo, fl, fl_type, wild_point_loc)
    print(f"Unique values after {fl_type}:", np.unique(result))
    result_img = Image.fromarray(result.astype(np.uint8))
    result_img.save(f'Results/{fl_type}.png')
    result_img.show()

print('Done')
```
